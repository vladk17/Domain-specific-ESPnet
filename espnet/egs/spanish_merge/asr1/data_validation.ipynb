```python
DATASET_PATH = 'gong'
GENERATIVE_PATH = f'data/train_{DATASET_PATH}'
```

```python
import os
import pandas as pd
import random
import IPython.display as ipd
```

```python
path_to_wav = os.path.join(GENERATIVE_PATH, 'wav.scp')
path_to_text = os.path.join(GENERATIVE_PATH, 'text')
```

```python
with open(path_to_text) as f:
    d = f.readlines()
texts = {item.split(' ')[0]: ' '.join(item.split(' ')[1:]).replace('\n','')  for item in d}
```

```python
with open(path_to_wav) as f:
    d = f.readlines()
wavs = {item.split(' ')[0]: ' '.join(item.split(' ')[1:]).replace('\n','')  for item in d}
```

```python
random_utterances = []
for i in range(10):
    utt_id = random.choice(list(texts.keys()))
    text = texts[utt_id]
    wav_path = wavs[utt_id]
    random_utterances.append({"id":utt_id,'text':text,'path':wav_path})
```

```python
utt = random.choice(random_utterances)
print('id:', utt['id'])
print('text:', utt['text'])
ipd.Audio(utt['path'])
```
